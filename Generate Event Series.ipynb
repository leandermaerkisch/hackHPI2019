```python
import pandas as pd
import numpy as np
import datetime
import json
import codecs
from progressbar import ProgressBar
```

# Data Preparation

```python
data = pd.read_csv('../2018-01-01-2019-12-31-Western_Africa.csv')
data.head(5)
```

```python
def to_datetime(datestring):
    months = {
        'January': '01',
        'February': '02',
        'March': '03',
        'April': '04',
        'May': '05',
        'June': '06',
        'July': '07',
        'August': '08',
        'September': '09',
        'October': '10',
        'November': '11',
        'December': '12',
    }
    parts = datestring.split()
    parts[1] = months[parts[1]]
    datestring = " ".join(parts)
    return datetime.datetime.strptime(datestring, "%d %m %Y").date()
```

```python
def get_actor_based_event_series(data):
    actor_dict = {}
    for actor in ProgressBar()(data['actor1'].unique()):
        actor_records = data[data['actor1'] == actor]
        actor_records['event_date'] = actor_records['event_date'].apply(lambda x: to_datetime(x))
        sorted_data = actor_records.sort_values(['event_date', 'event_id_cnty', 'time_precision'])
        actor_dict[actor] = sorted_data
    return actor_dict
```

```python
actor_frames = get_actor_based_event_series(pd.read_csv('../2018-01-01-2019-12-31-Western_Africa.csv'))
```

```python
actor_frames['Protesters (Burkina Faso)']
```

# Deriving Month Centroids

```python
def get_month_centroids(data):
    centroids = []
    for y in range(2018, 2020):
        for m in range(1, 13):
            month_points = [data.iloc[i]  for i in range(len(data)) if data.iloc[i]['event_date'].month == m and data.iloc[i]['event_date'].year == y]
            if month_points:
                mframe = pd.DataFrame.from_records(month_points)
                centroid = [np.mean(mframe['latitude']), np.mean(mframe['longitude'])]
                m_object = {}
                m_object['date'] = '-'.join([str(y), str(m)])
                m_object['latlng'] = centroid
                mframe['event_date'] = mframe['event_date'].apply(lambda x: str(x))
                m_object['incidents'] = json.loads(mframe.to_json(orient='records'))
                centroids.append(m_object)
    return centroids
```

```python
month_centroids = {}
for actor, frame in ProgressBar()(actor_frames.items()):
    month_centroids[actor] = get_month_centroids(frame)
```

```python
json_file = "month_centroids_per_actor.json" 
json.dump(month_centroids, codecs.open(json_file, 'w', encoding='utf-8'), sort_keys=True, indent=4)
```

# Gather Border Crossings

```python
countries = {}
for i in ProgressBar()(range(len(data) - 1)):
    actor = data.iloc[i]['actor1']
    if actor in countries:
        countries[actor].add(data.iloc[i]['country'])
    else:
        countries[actor] = set([data.iloc[i]['country']])
relevant_actors = {}
for actor, country in countries.items():
    if len(country) > 1:
        relevant_actors[actor] = country
relevant_borders = {}
for actor, country_set in relevant_actors.items():
    for country1 in country_set:
        for country2 in country_set:
            if country1 >= country2:
                continue
            if country1 < country2:
                if actor in relevant_borders:
                    relevant_borders[actor].append(', '.join([country1, country2]))
                else:
                    relevant_borders[actor] = [', '.join([country1, country2])]
relevant_borders
```

```python
json_file = "crossed_borders_per_actor.json" 
json.dump(relevant_borders, codecs.open(json_file, 'w', encoding='utf-8'), sort_keys=True, indent=4)
```

# Rank consecutive pairs of actors

```python
pairs = {}
for i in ProgressBar()(range(len(data) - 1)):
    hunter = data.iloc[i]['actor1']
    hunted = data.iloc[i+1]['actor1']
    if hunter == hunted:
        continue
    key = ' --> '.join([hunter, hunted])
    if key in pairs:
        pairs[key] = pairs[key] + 1
    else:
        pairs[key] = 1
pair_frame = pd.DataFrame()
pair_frame['relation'] = list(pairs.keys())
pair_frame['frequency'] = list(pairs.values())
pair_frame = pair_frame.sort_values('frequency', ascending=False).reset_index(drop=True)
pair_frame
```

```python
json_pairs = json.loads(pair_frame.to_json(orient='records'))
json_file = "consecutive_actors_frequency_ranking.json" 
json.dump(json_pairs, codecs.open(json_file, 'w', encoding='utf-8'), sort_keys=True, indent=4)
```
